## 縣市

```python
import pandas as pd

df = pd.read_csv('C:\lab\\aigo\\30_Private Dataset _Private and Publict Submission Template_v2\public_private_dataset_process.csv')
price = pd.read_csv('C:\lab\\aigo\\30_Training Dataset_V2\price_for_city.csv')

print(df.columns)
print(df.shape)
print(price.columns)
print(price.shape)
```

```python
price_city = price[['縣市','縣市最低單價', '縣市最高單價', '縣市平均單價', '縣市單價中位數']].drop_duplicates()
print(price_city.shape)
price_city
```

```python
df = df.merge(price_city, on=['縣市'], how='left')
df
```

```python
df.to_csv('C:\lab\\aigo\\30_Private Dataset _Private and Publict Submission Template_v2\public_private_dataset_process_with_price.csv', index=False)
```

## 鄉鎮市區

```python
import pandas as pd
chunksize = 1000

df = pd.read_csv('C:\lab\\aigo\\30_Private Dataset _Private and Publict Submission Template_v2\public_private_dataset_process_with_price.csv')
# data_iterator = pd.read_csv('C:\lab\\aigo\\30_Private Dataset _Private and Publict Submission Template_v2\public_private_dataset_process_with_price.csv', chunksize=chunksize)
price = pd.read_csv('C:\lab\\aigo\\30_Training Dataset_V2\price_for_city.csv')

print(df.columns)
print(df.shape)
print(price.columns)
print(price.shape)
```

```python
price_area = price[['縣市','鄉鎮市區', '鄉鎮市區最低單價','鄉鎮市區最高單價', '鄉鎮市區平均單價', '鄉鎮市區單價中位數']].drop_duplicates()
print(price_area.shape)
price_area
```

```python
df = df.merge(price_area, on=['縣市','鄉鎮市區'], how='left')
df
```

```python
# 填充缺失值
df['鄉鎮市區最低單價'].fillna(df['縣市最低單價'], inplace=True)
df['鄉鎮市區最高單價'].fillna(df['縣市最高單價'], inplace=True)
df['鄉鎮市區平均單價'].fillna(df['縣市平均單價'], inplace=True)
df['鄉鎮市區單價中位數'].fillna(df['縣市單價中位數'], inplace=True)
```

```python
df
```

```python
df.to_csv('C:\lab\\aigo\\30_Private Dataset _Private and Publict Submission Template_v2\public_private_dataset_process_with_price.csv', index=False)
```

```python
import pandas as pd
chunksize = 1000

df = pd.read_csv('C:\lab\\aigo\\30_Private Dataset _Private and Publict Submission Template_v2\public_private_dataset_process_with_price.csv')
price = pd.read_csv('C:\lab\\aigo\\30_Training Dataset_V2\price_for_city.csv')

print(df.columns)
print(df.shape)
print(price.columns)
print(price.shape)
```

```python
price_road = price[['縣市','鄉鎮市區','路名', '該路段最低單價', '該路段最高單價', '該路段平均單價','該路段單價中位數']].drop_duplicates()
print(price_road.shape)
price_road
```
